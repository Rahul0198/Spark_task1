```python
#Importing all required libraries for prediction and analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#Importing required data from server
link = "http://bit.ly/w-data"
data = pd.read_csv(link)
print("recieved the data and displayed a preview below")
data.head()
```

```python
#Plotting the scatter plot of recieved  data 
data.plot(x = 'Hours', y = 'Scores', style='.',color='r')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied By Student')  
plt.ylabel('Percentage Obtained')  
plt.show()
```

```python
#training and testing sets from 'Data'
X = data.iloc[:,:-1].values
y = data.iloc[:,1].values
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=0)
```

```python
#Training the 'Data'
reg = LinearRegression()
reg.fit(X_train, y_train)
print("The training is completed")
```

```python
#Plotting for the test data
st_line = reg.coef_*X+reg.intercept_
plt.scatter(X,y)
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied By Student')  
plt.ylabel('Percentage Obtained') 
plt.plot(X,st_line,'r');
plt.show()
```

```python
#Predicting using input X_test which we had splitted
print(X_test)
y_pred = reg.predict(X_test)
```

```python
#Comparing Actual vs Predicted Data
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
print(df)
```

```python
#Comparing by using graphical means
plt.scatter(X_test,y_pred,marker='^',c='r')
plt.scatter(X_test,y_test,c='c')
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied By Student')  
plt.ylabel('Percentage Obtained') 
plt.show
```

```python
#Question : score of a student who studies 9.25 hrs/day
time = [[9.25]]
Obtained = reg.predict(time)
print("Obtained score out of 100=",Obtained)
```

```python
#Accuracy of Model
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred), "\n")
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred), "\n")
print('R-squared:', metrics.r2_score(y_test, y_pred)) 
```
